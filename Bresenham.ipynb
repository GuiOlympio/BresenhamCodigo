```python
x1 = int(input("X inicial: "))
x2 = int(input("X Final: "))

y1 = int(input("Y inicial: "))
y2 = int(input("Y Final: "))

dx = x2 - x1
dy = y2 - y1

pd = 2*dy - dx

x = x1
y = y1

print(f"({x}, {y})")
while True:
  if pd < 0:
    x += 1

    pd += 2*dy

    print(f"({x}, {y})")

  elif pd >= 0:
    x += 1
    y += 1

    pd += 2*(dy - dx)

    print(f"({x}, {y})")

  if x == x2 and y == y2:
    break
```
